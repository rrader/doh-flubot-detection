 - inspect the AUC curve of top models
 - select the most interesting thresholds
 - calculate the confusion matricies for them

```python
#| output: false
p1version = "v6"
p2version = "v7"
p3version = "v5"
```

```python
test = pd.read_feather(f'cached/robustness.balanced_mix.p3.{p3version}.ft')
```

```python
# results_df = pd.read_feather(f"cached/robustness-flubot-mix.p3.{p3version}.ft")
results_df = pd.read_feather(f"cached/robustness.balanced_mix-results.p3.{p3version}.ft")
results_df["skips_and_weights"] = results_df["skips_and_weights"].apply(str)
```

```python
best_normed_models = results_df[results_df["normed"] == True].sort_values("test_data_auc", ascending=False).iloc[:3]
best_unnormed_models = results_df[results_df["normed"] == False].sort_values("test_data_auc", ascending=False).iloc[:3]
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from joblib import dump, load


sensitivities = [0.4, 0.6, 0.8, 0.9]
check_thresholds = [0.69, 0.5, 0.18]


for m in [best_normed_models.iloc[0], best_unnormed_models.iloc[0]]:

    model, normed, skips_and_weights, auc, fpr, tpr, thresholds = m
    skips_and_weights = [int(i) for i in skips_and_weights.strip("[]").split()]
    name = (
        ('normed-' if normed else 'unnormed-') + 
        '-'.join(str(p) for p in skips_and_weights)
    )
    clf = load(f'models/{model}-mix-1-p-{name}.p1.{p1version}.p2.{p2version}.joblib')

    pp = PreProcessing(normed, *skips_and_weights)
    full, features, labels = pp.process(test, f"cached/robustness.balanced_mix.p3.{p3version}.{name}.saved")

    plt.plot(fpr,tpr,label=f"{name} AUC={auc}", color='red')

    plt.legend()
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.show()

    fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(12, 8))
    axs = np.ravel(axs)

    for i, sensitivity in enumerate(sensitivities):
        thres = [thres1 for sens1, thres1 in zip(tpr, thresholds) if sens1 >= sensitivity][0]

        print("sensitivity:", sensitivity, "threshold:", thres)
        y_pred = (clf.predict_proba(features[pp.feature_fields])[:,1] >= thres).astype(bool)
        cm = confusion_matrix(labels, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['NonDoH', 'DoH'])
        disp.plot(ax=axs[i], colorbar=False)
        f1 = metrics.f1_score(labels, y_pred)
        axs[i].set_title(f"thres={thres} tpr={sensitivity} ({model} {name})\nf1={f1:.2f}")

    fig.tight_layout()
    display(pd.DataFrame({"tpr": tpr[::15], "fpr": fpr[::15], "threshold": thresholds[::15]}).T)
    plt.show()
```
