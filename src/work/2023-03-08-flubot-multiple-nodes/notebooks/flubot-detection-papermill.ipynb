```python
import papermill as pm
```

```python
results_df = pd.read_feather(f"cached/robustness.balanced_mix-results.p3.{p3version}.ft")
```

```python
models_0_0_0_0_normed = results_df[(results_df["normed"] == True) & (results_df["skips_and_weights"].apply(str) == '[0 0 0 0]')].sort_values("test_data_auc", ascending=False).iloc[:1]
models_0_0_0_0_unnormed = results_df[(results_df["normed"] == False) & (results_df["skips_and_weights"].apply(str) == '[0 0 0 0]')].sort_values("test_data_auc", ascending=False).iloc[:1]
best_normed_models = results_df[results_df["normed"] == True].sort_values("test_data_auc", ascending=False).iloc[:3]
best_unnormed_models = results_df[results_df["normed"] == False].sort_values("test_data_auc", ascending=False).iloc[:3]
```

```python
import nbformat as nbf
```

```python
import json
import nbformat as nbf

os.makedirs(f'./flubot-detection-try-models', exist_ok=True)

for _, m in pd.concat([
    models_0_0_0_0_normed, models_0_0_0_0_unnormed, 
    best_normed_models, best_unnormed_models
]).iterrows():
    model, normed, skips_and_weights, auc, fpr, tpr, thresholds = m
    name = (
        ('normed-' if normed else 'unnormed-') + 
        '-'.join(str(p) for p in skips_and_weights)
    )
    print(model, name)
    output = f'./flubot-detection-try-models/gen-flubot-detection-{model}-{name}.ipynb'
    pm.execute_notebook(
       f'./flubot-detection-one.ipynb',
       output,
       parameters=dict(
           model=model,
           normed=normed,
           skips_and_weights=skips_and_weights.tolist()
       )
    )
    ntbk = nbf.read(output, nbf.NO_CONVERT)
    for cell in ntbk.cells:
        if 'title' in cell.metadata.tags:
            cell.source = f"# Detect malware ({model} {name})"
    nbf.write(ntbk, output, nbf.NO_CONVERT)
```
